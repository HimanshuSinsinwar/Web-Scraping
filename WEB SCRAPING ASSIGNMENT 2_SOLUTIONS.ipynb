### python program to scrape data for “Data Analyst” Job position in“Bangalore” location.

```python
browser.get("https://www.naukri.com/")

time.sleep(2)
##Entering “Data Analyst” in “Skill,Designations,Companies” field
title='data analyst'
job_title=browser.find_elements_by_id('qsb-keyword-sugg')
job_title[0].send_keys(title)
##Entering “Bangalore” in “enter the location” field.
job_location=browser.find_elements_by_id('qsb-location-sugg')
location='banglore'
job_location[0].send_keys(location)
##clicking
search_button=browser.find_elements_by_xpath("//button[@class='btn']")
search_button[0].click()

Job_details=browser.find_elements_by_xpath("//article[@class='jobTuple bgWhite br4 mb-8']")
job_data=[]
n=0
for a in Job_details:
        if n<11:
            n+=1
            job_data.append(a.text)
        else:
            pass

        raw_data=pd.DataFrame(job_data)[0].str.split('\n', n=-1, expand=True)
raw_data
```
