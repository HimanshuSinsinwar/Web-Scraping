```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
from selenium import webdriver
```

```python
driver_path=r'C:\Users\ADMIN\Documents\chromedriver.exe'
```

```python
browser = webdriver.Chrome(executable_path=driver_path)
```

```python
browser.get("https://www.imdb.com/search/title/?groups=top_100&view=advanced")
```

```python
rating_ids=browser.find_elements_by_xpath("//strong")
name_ids=browser.find_elements_by_xpath("//a[contains(@href,'?ref_=adv_li_tt')]")
year_id=browser.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']")
```

```python
rating=[]
for m in rating_ids:
    n=m.text
    if (n!="Detailed") and (n!="Popularity"):
        rating.append(n)
```

```python
name=[]
for m in name_ids:
    n=m.text
    name.append(n)
```

```python
year=[]
for m in year_id:
    n=m.text
    year.append(n)
```

```python
imdb1_data=pd.DataFrame({'Name':name,'Rating':rating,'Year':year})
imdb1_data
```
